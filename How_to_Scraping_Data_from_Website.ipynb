```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://en.wikipedia.org/wiki/List_of_largest_companies_in_Indonesia"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
table = soup.find_all('table', class_= 'wikitable sortable')[1]
```

```python
table_headers = table.findAll('th')
```

```python
list_header = [header.text.strip() for header in table_headers]
```

```python
import pandas as pd
df = pd.DataFrame(columns=list_header)
```

```python
df
```

```python
column_data = table.findAll('tr')
print(column_data
)
```

```python
for row in column_data[1:]:
  row_data = row.findAll('td') # row data akan menyimpan data
  ind_row_data = [data.text.strip() for data in row_data]

  #masukkan data untuk tiap cell keadalam df
  length = len(df)
  df.loc[length] = ind_row_data
```
